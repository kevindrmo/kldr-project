# **Question 2: The Relationship Between Digital Adoption and Digital Exports**

This question investigates the causal impact of a country's digital adoption on its volume of digital service exports.

**Our Core Challenge:**
1.  **Endogeneity (Reverse Causality):** Does adoption drive exports, or do exports drive adoption?
2.  **Omitted Variable Bias (OVB):** Are unobserved factors (like stable governance) driving both?

**Our Strategy:**
1.  **Step 1: Data Assembly.** We will build a complete panel dataset by fetching control variables (GDP, Population) and a proxy for digital adoption (Internet Usage %) from the World Bank API and merging it with our existing digital exports data.
2.  **Step 2: Baseline Model (OLS).** We will run a simple OLS regression to establish a baseline correlation, and we will discuss why this model is likely biased.
3.  **Step 3: Robust Model (Fixed Effects).** We will use a Panel Data model with Country Fixed Effects to control for all time-invariant, country-specific omitted variables.
4.  **Step 4: Advanced Model (Instrumental Variable).** We will attempt to find a valid instrument to address endogeneity and estimate the causal effect using a Two-Stage Least Squares (2SLS) model.


## **Step 1.1: Data Acquisition from World Bank API**

Acquire External Control Variables (GDP & Population)**

Our dataset is missing GDP and Population. We will download these two variables from the World Bank API. This is the only external data we need.

We will download this data for all countries from 2010-2023 and save it to a clean intermediate CSV file.


```python
import wbdata
import pandas as pd
from datetime import datetime

# --- Define only the missing indicators ---
indicators = {
    "NY.GDP.PCAP.KD": "gdp_per_capita",
    "SP.POP.TOTL": "population"
}
data_date = (datetime(2010, 1, 1), datetime(2023, 1, 1))

# --- Download data ---
print("Downloading GDP and Population data from World Bank API...")
wb_controls = wbdata.get_dataframe(indicators, country="all", date=data_date)
print("Download complete.")


# --- CORRECTED CLEANING AND PREPARATION ---

# 1. Turn the index (country, date) into regular columns
wb_cleaned = wb_controls.reset_index()

# 2. Rename the columns for consistency
wb_cleaned.rename(columns={'date': 'year', 'country': 'country_name'}, inplace=True)

# 3. Convert year to a number
wb_cleaned['year'] = pd.to_numeric(wb_cleaned['year'])

# 4. Now, create the 'country_code' column from the 'country_name' column
country_mapping = wbdata.get_countries()
country_codes = {item['name']: item['id'] for item in country_mapping}
wb_cleaned['country_code'] = wb_cleaned['country_name'].map(country_codes)

# 5. Drop non-country aggregate rows (like "Africa Eastern and Southern")
wb_cleaned.dropna(subset=['country_code'], inplace=True)

# 6. Save the properly cleaned data to a CSV
wb_cleaned.to_csv("world_bank_cleaned_controls.csv", index=False)

print("\nCleaned control variables (with country_code) saved to 'world_bank_cleaned_controls.csv'")
wb_cleaned.head(50)


```

```python
wb_cleaned.to_csv(OUT_CLEAN/"world_bank_cleaned_controls.csv", index=False)
```

## Now I am about to create my regression model

```python
print(OUT_CLEAN)
```

```python
import statsmodels.api as sm
```

```python
df = pd.read_csv(OUT_CLEAN/"question1"/"final_panel_for_regression.csv")

# --- Prepare variables for the regression ---
# 1. Define the dependent variable (Y) and apply a log transformation
#    (Using your original column name for clarity)

Y= np.log(df["Exports_Digital_Service"] +1)

# 2. Define the independent variables (X)
#    (Using the new, clean column name from the World Bank)

X = df[["internet_usage_pct", "gdp_per_capita", "population"]]

# 3. Apply log transformations to the control variables

X['log_gdp_per_capita'] = np.log(X['gdp_per_capita'] + 1)
X['log_population'] = np.log(X['population'] + 1)

# 4. Select the final set of regressors for the model
X = X[['internet_usage_pct', 'log_gdp_per_capita', 'log_population']]

# 5. Add a constant (the intercept, β₀) to the model
X = sm.add_constant(X)

# --- Run the OLS model ---

model_ols = sm.OLS(Y, X)
results_ols = model_ols.fit()

# --- Print the summary table ---
print("--- OLS Regression Results ---")
print(results_ols.summary())
```
