```python
# df["indicator_name"].map(new_indicator_value_name_map, inplace=True) --> won't work, we have to assign it to the column

df["indicator_name"] = df["indicator_name"].map(new_indicator_value_name_map)
```

## Now we are ready to melt our dataframe into long format!

# Little validity check: create dataset with top exporters (lets say top 10)

```python
df_exports = df_long_clean[df_long_clean["indicator_name"] == "Exports_Digital_Service"]

df_exports.head(5)
```

```python
total_exports_by_country = df_exports.groupby("country_name")["value"].sum()
total_exports_by_country
```

```python
top_10_countries = total_exports_by_country.nlargest(10)
top_10_countries
```

```python
# filter out world aggregation

top_10_countries_without_world = top_10_countries.drop("World")
# we have a problem: trailing .0

```

```python
top_10_countries_without_world = top_10_countries_without_world.round(0).astype(int)
top_10_countries_without_world

```

```python
# Export the table:

out_path_top10_exporters = OUT_CLEAN/ "table_top10_exporters.csv"

top_10_countries_without_world.to_csv(out_path_top10_exporters)
```

# Using Pandas styler to make the table nicer for display

```python
table_top_10 = top_10_countries_without_world.reset_index()

table_top_10.columns = ["Country", "Total Digital Exports (Millions USD)"]

table_top_10["Rank"] = table_top_10.index +1

table_top_10 = table_top_10[["Rank", "Country", "Total Digital Exports (Millions USD)"]]

table_top_10


out_path_top10_exporters = OUT_CLEAN/ "table_top10_exporters.csv"

table_top_10.to_csv(out_path_top10_exporters, index=False)
```

```python


# --- 1. DEFINE STYLING (The CSS rules) ---
# NOTE: This styling must be defined first to be applied below.
custom_styles = [
    # Table background and borders
    {'selector': 'table',
     'props': [('background-color', '#1a1a2e'), ('color', '#cccccc'), ('border', '1px solid #ff8800'), ('border-radius', '12px'), ('overflow', 'hidden')]},
    # Header styling
    {'selector': 'th',
     'props': [('background-color', '#FF8C00'), ('color', '#000000'), ('font-weight', '700'), ('text-align', 'center'), ('padding', '10px')]},
    # Data cell styling
    {'selector': 'td',
     'props': [('padding', '8px 12px'), ('border-bottom', '1px solid #0f3460'), ('text-align', 'center')]},
    # Highlight the Rank 1 row
    {'selector': 'tr:first-child',
     'props': [('font-weight', 'bold'), ('background-color', '#404058')]},
]

# --- 2. EXECUTE SAVE TO HTML FILE ---



# Define the file path
html_path = OUT_CLEAN/ "interactive_styled_exporters.html"

# Apply styling and save to the HTML file
styled_output = (
    table_top_10.style
    .hide(axis='index')
    .format({'Total Digital Exports (Millions USD)': lambda x: f'${x:,.0f}'})
    .set_caption("Top 9 Digital Service Exporters (2010-2023 Total)")
    .set_table_styles(custom_styles)
)

# SAVE COMMAND
styled_output.to_html(html_path)

print(f"--- SUCCESS ---")
print(f"Your styled table is saved to: {html_path}")
print("You can now open this file locally and insert the HTML content into your application.")
```
